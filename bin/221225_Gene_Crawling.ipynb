```python
def digest(_str, keyword):
    l_ = []
    
    s = ''
    flag = -1
    s_1 = ''
    s_2 = ''
    for char in _str:
        s = s + char
        
        ## closing
        if flag == 1 and char == ',' or flag == 1 and char == '.':
            flag = -1
            l_.append(s_1)
            s_1 = ''
        elif flag == 2 and char == ')':
            flag = -1
            l_.append(s_2)
            s_2 = ''

        ## propagation
        elif flag == 1:
            s_1 = s_1 + char
        elif flag == 2:
            s_2 = s_2 + char
            
        ## tigger
        elif s.endswith(keyword + ', '):
            flag = 1
        elif s.endswith(keyword + ' ('):
            flag = 2

    return l_
```
